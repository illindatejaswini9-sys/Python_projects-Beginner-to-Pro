```python
# Create a Contact_Book

#1. Initialize empty contact book

Contacts = {}

#2 create menu
def show_menu():
  print("\nContact Book Menu:")
  print("1. Add Contact")
  print("2. Search Contact")
  print("3. Edit Contact")
  print("4. Delete Contact")
  print("5. Show All Contacts")
  print("6. Quit")

#3 Add contact
def add_contact():
  name = input("Enter Contact Name: ")
  number = input("Enter Phone Number: ")
  email = input("Enter Email: ")

  Contacts[name] = {"Number": number, "Email": email}
  print("Contact added successfully!")
  input_choice = input("Do you want to add another contact? (yes/no): ")
  if input_choice.lower() == "yes":
   add_contact()
  else:
    show_menu()

#4 Search contact
def search_contact():
  name = input("Enter Contact Name you want to search: ")
  if name in Contacts:
    contact = Contacts[name]
    print(f"Name: {name}")
    print(f"Number: {contact['Number']}")
    print(f"Email: {contact['Email']}")
  else:
    print(f"Contact {name} not found.")

#5 Edit Contact
def edit_contact():
  name = input("Enter name you want to edit:")
  if name in Contacts:
    phone = input("Enter new phone number: ")
    email = input("Enter new email: ")
    Contacts[name] = {"Number": phone, "Email": email}
    print("Contact updated successfully!")
  else:
    print(f"Contact {name} not found.")

#6 Delete Contact
def delete_contact():
  name = input("Enter name you want to delete: ")
  if name in Contacts:
    del Contacts[name]
    print("Contact successfully deleted")
  else:
    print("Contact not found")

#7 Show all contacts
def view_contact():
 if Contacts:
  print("\n--All Contacts:--")
  for name,details in Contacts.items():
    print(f"Name: {name}")
    print(f"Number: {details['Number']}")
    print(f"Email: {details['Email']}")
 else:
    print("No contacts found.")

  #8 Main programm loop
while True:
    show_menu()
    choice = input("Enter your choice(1-6)")

    if choice == "1":
      add_contact()
    elif choice =="2":
      search_contact()
    elif choice =="3":
      edit_contact()
    elif choice == "4":
      delete_contact()
    elif choice =="5":
      view_contact()
    elif choice =="6":
      print("Thank you for using the Contact Book.Good Bye!")
      break
    else:
      print("Invalid choice.Please select a valid option(1-6).")
```
