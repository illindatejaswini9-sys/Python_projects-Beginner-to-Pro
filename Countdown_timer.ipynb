```python
#countdown timer using for loop
import time
#step 1 :Get user input for countdown start

start = int(input("give countdown time for the timer:"))

for i in range(start,0,-1):
  time.sleep(1)
  print(i)
```

```python
#countdown timer using while loop
import time

start = int(input("Give Countdown time for the timer:"))
speed = int(input("Give speed of the countdown timer less than time:"))

print("\n --Countdown begins--")
while start>0 and speed<start:
  time.sleep(speed)
  print(start)
  start -=1
print("Countdown end")

if(speed>start):
  print("Speed is greater than time")

```
